```python
# ============================================
# Notebook 07: Feature Engineering
# Purpose: Encode categorical variables and
# prepare final feature matrix for modeling
# ============================================

# Mount Google Drive so we can access our raw data files
from google.colab import drive
drive.mount('/content/drive')

# Standard imports
import os
import pandas as pd

# Define folder paths so we don't repeat long strings everywhere
# RAW = where original untouched downloads live
# CLEANED = where filtered/cleaned data gets saved
# MERGED = where the final combined dataset goes (used in notebook 05)
RAW = '/content/drive/MyDrive/rgv_business_survival/data/raw'
CLEANED = '/content/drive/MyDrive/rgv_business_survival/data/cleaned'
MERGED = '/content/drive/MyDrive/rgv_business_survival/data/merged'

# Make sure cleaned folder exists
os.makedirs(CLEANED, exist_ok=True)

# Verify our raw files are accessible
print("Merged files:", sorted(os.listdir(MERGED)))

model_df = pd.read_csv(f'{MERGED}/model_df.csv')
print(f'Loaded: {model_df.shape}')
```

```python
# Drop columns that are identifiers or already captured elsewhere
# borrname, borrcity = just names, not predictive
# loanstatus = already captured in 'defaulted' column
# naicscode = already captured in 'naicsdescription'
# projectcounty = too many categories, zip code already captures location
# year, approvalfiscalyear = same column duplicated after merge

drop_cols = ['borrname', 'borrcity', 'loanstatus', 'naicscode',
             'projectcounty', 'approvalfiscalyear']

model_df = model_df.drop(columns=drop_cols)
print(f'After dropping: {model_df.shape}')
print(model_df.columns.tolist())
```

```python
# Encode categorical columns using one-hot encoding
# get_dummies creates a new binary column for each unique value
# drop_first=True removes one category to avoid multicollinearity
cat_cols = ['naicsdescription', 'businesstype', 'businessage']

model_df = pd.get_dummies(model_df, columns=cat_cols, drop_first=True)

print(f'After encoding: {model_df.shape}')
print(f'New columns added: {model_df.shape[1] - 16}')
```

```python
# Save the final feature matrix to merged folder
# This is what notebook 08 will load to train the model
model_df.to_csv(f'{MERGED}/model_features.csv', index=False)
print(f'Saved: {model_df.shape[0]} rows, {model_df.shape[1]} columns')
print(os.listdir(MERGED))
```
